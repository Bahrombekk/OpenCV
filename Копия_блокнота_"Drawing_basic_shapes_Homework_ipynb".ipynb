```python
def face_bounding_box(img):
  '''
  Face bounding box
  '''
  # CODE
  a=cv.rectangle(img,(130,30),(600,650),(0,0,255),thickness=5)
  a=cv.putText(img,'Elon Musk',(110,800),16,3,(0,0,255),thickness=5)
  b=cv.rectangle(img,(930,30),(1400,650),(0,0,255),thickness=5)
  b=cv.putText(img,'Jeff Bezos',(900,800),16,3,(0,0,255),thickness=5)
  plt.imshow(img[:,:,[2,1,0]],)
  plt.show()
img=cv.imread('muskbezos.jpg')
face_bounding_box(img)
```

```python
def bounding_box1(img,start_point,end_point,color):
  '''
  Draw bounding box
  '''
  # CODE
  pts=np.array([[200,200],
                [200,800],
                [800,800],
                [800,200]])
  pts1=np.array([[200,200],
                 [240,200],
                 [240,240],
                 [200,240]])
  pts2=np.array([[800,200],
                 [760,200],
                 [760,240],
                 [800,240]])
  pts3=np.array([[200,800],
                 [200,760],
                 [240,760],
                 [240,800]])
  pts4=np.array([[800,800],
                 [760,800],
                 [760,760],
                 [800,760]])
  a=cv.polylines(img,[pts],True,(0,0,0),thickness=1)
  b=cv.polylines(img,[pts1],True,(0,0,255),thickness=8)
  b1=cv.polylines(img,[pts2],True,(0,0,255),thickness=8)
  b2=cv.polylines(img,[pts3],True,(0,0,255),thickness=8)
  b3=cv.polylines(img,[pts4],True,(0,0,255),thickness=8)
  c=cv.fillPoly(img,[pts],(255,255,255))
  plt.imshow(img)
  plt.show()

img = cv.imread('image_draw.png')
bounding_box1(img,(200,200),(800,800),(0, 0, 255))
```

```python
def bounding_box2(img,start_point,end_point,color):
  '''
  Draw bounding box
  '''
  # CODE
  pts=np.array([[220,220],
                [220,780],
                [780,780],
                [780,220]])
  pts1=np.array([[200,200],
                 [240,200],
                 [240,240],
                 [200,240]])
  pts2=np.array([[800,200],
                 [760,200],
                 [760,240],
                 [800,240]])
  pts3=np.array([[200,800],
                 [200,760],
                 [240,760],
                 [240,800]])
  pts4=np.array([[800,800],
                 [760,800],
                 [760,760],
                 [800,760]])
  a=cv.polylines(img,[pts],True,(255,0,0),thickness=10)
  b=cv.polylines(img,[pts1],True,(255,0,0),thickness=10)
  b1=cv.polylines(img,[pts2],True,(255,0,0),thickness=10)
  b2=cv.polylines(img,[pts3],True,(255,0,0),thickness=10)
  b3=cv.polylines(img,[pts4],True,(255,0,0),thickness=10)
  c1=cv.fillPoly(img,[pts1],(255,255,255))
  c2=cv.fillPoly(img,[pts2],(255,255,255))
  c3=cv.fillPoly(img,[pts3],(255,255,255))
  c4=cv.fillPoly(img,[pts4],(255,255,255))
  plt.imshow(img)
  plt.show()

img = cv.imread('image_draw.png')
bounding_box2(img,(200,200),(800,800),(255, 0, 0))
```

```python
def bounding_box3(img,start_point,end_point,color):
  '''
  Draw bounding box
  '''
  # CODE
  pts=np.array([[200,200],
                [200,800],
                [800,800],
                [800,200]])
  pts1=np.array([[200,200],
                 [240,200],
                 [240,240],
                 [200,240]])
  pts2=np.array([[800,200],
                 [760,200],
                 [760,240],
                 [800,240]])
  pts3=np.array([[200,800],
                 [200,760],
                 [240,760],
                 [240,800]])
  pts4=np.array([[800,800],
                 [760,800],
                 [760,760],
                 [800,760]])
  a=cv.polylines(img,[pts],True,(0,0,0),thickness=1)
  b=cv.polylines(img,[pts1],True,(0,0,0),thickness=3)
  b1=cv.polylines(img,[pts2],True,(0,0,0),thickness=3)
  b2=cv.polylines(img,[pts3],True,(0,0,0),thickness=3)
  b3=cv.polylines(img,[pts4],True,(0,0,0),thickness=3)
  c=cv.fillPoly(img,[pts],(250, 248, 235))
  c1=cv.fillPoly(img,[pts1],(255,0,0))
  c2=cv.fillPoly(img,[pts2],(255,0,0))
  c3=cv.fillPoly(img,[pts3],(255,0,0))
  c4=cv.fillPoly(img,[pts4],(255,0,0))
  plt.imshow(img)
  plt.show()

img = cv.imread('image_draw.png')
bounding_box3(img,(200,200),(800,800),(227, 220, 220))
```

```python
def bounding_box4(img,start_point,end_point,color):
  '''
  Draw bounding box
  '''
  # CODE
  c=cv.rectangle(img,(200,300),(800,700),(255,0,0),thickness=5)
  b=cv.circle(img,(200,300),40,(255,0,0),-1)
  b1=cv.circle(img,(200,700),40,(255,0,0),-1)
  b2=cv.circle(img,(800,700),40,(255,0,0),-1)
  b3=cv.circle(img,(800,300),40,(255,0,0),-1)
  plt.imshow(img)
  plt.show()

img = cv.imread('image_draw.png')
bounding_box4(img,(200,300),(800,700),(255, 0, 0))
```

```python
def draw_triangle(img):
  '''
  Drawing triangle
  '''
  # CODE
  pts=np.array([[450,200],
                [200,800],
                [800,800]])
  a=cv.polylines(img,[pts],True,(0,0,0),thickness=5)
  plt.imshow(img)
  plt.show()

img = cv.imread('image_draw.png')
draw_triangle(img)
```

```python
def draw_rhombus(img):
  '''
  Drawing rhombus
  '''
  # CODE
  pts=np.array([[500,200],
       [200,500],
       [500,800],
       [800,500]])
  a=cv.polylines(img,[pts],True,(0,0,0),thickness=5)
  plt.imshow(img)
  plt.show()

img = cv.imread('image_draw.png')
draw_rhombus(img)
```

```python
def draw_star(img):
  '''
  Drawing star
  '''
  # CODE
  pts=np.array([[500,200],
             [200,650],
             [800,650]])
  pts1=np.array([[200,350],
                 [500,800],
                 [800,350]])
  a=cv.polylines(img,[pts],True,(0,0,0),thickness=5)
  b=cv.polylines(img,[pts1],True,(0,0,0),thickness=5)
  plt.imshow(img)
  plt.show()

img = cv.imread('image_draw.png')
draw_star(img)
```

```python
def windows_logo(img):
  '''
  Drawing windows logo
  '''
  # CODE
  fig=np.array([[0,0],
                [1000,0],
                [1000,1000],
                [0,1000]])
  pts=np.array([[200,200],
                [490,200],
                [490,490],
                [200,490]])
  pts1=np.array([[200,510],
                 [490,510],
                 [490,800],
                 [200,800]])
  pts2=np.array([[510,200],
                 [800,200],
                 [800,490],
                 [510,490]])
  pts3=np.array([[510,510],
                 [800,510],
                 [800,800],
                 [510,800]])
  c=cv.polylines(img,[fig],True,(0,0,0),thickness=5)
  c1=cv.fillPoly(c,[fig],(26, 100, 237))
  a=cv.polylines(img,[pts],True,(0,0,0),thickness=5)
  a1=cv.polylines(img,[pts1],True,(0,0,0),thickness=5)
  a2=cv.polylines(img,[pts2],True,(0,0,0),thickness=5)
  a3=cv.polylines(img,[pts3],True,(0,0,0),thickness=5)
  b1=cv.fillPoly(a,[pts],(207, 141, 19))
  b2=cv.fillPoly(a1,[pts1],(22, 161, 29))
  b3=cv.fillPoly(a2,[pts2],(28, 186, 214))
  b4=cv.fillPoly(a3,[pts3],(237, 220, 26))
  plt.imshow(img)
  plt.show

img = cv.imread('image_draw.png')
windows_logo(img)
```
