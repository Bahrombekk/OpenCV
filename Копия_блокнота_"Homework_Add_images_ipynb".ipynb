```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img=cv2.imread('lena.jpg')
```

```python
zeros = np.zeros(img.shape,dtype= np.uint8)
plt.imshow(zeros)
plt.show()
```

```python
def full_bbox_image(center_x,center_y,width,height,img):
  # Input Format: YOLO
  # CODE
  img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  zeros = np.zeros(img.shape,dtype= np.uint8)
  zeros=cv2.rectangle(zeros,(center_x-width//2,center_y-height//2),(center_x+width//2,center_y+height//2),(255,0,0),-1)
  plt.imshow(cv2.addWeighted(zeros,0.3,img,0.9,0))
  plt.show()

img = cv2.imread('lena.jpg')
full_bbox_image(150,150,100,100,img)
```

```python
img = cv2.imread('lena.jpg')
full_bbox_image(140,200,90,100,img)
```

```python
def add_logo(center_x,center_y,img_logo):

  # CODE
  h,w,d=img_logo.shape
  img_z=np.zeros((700,700,3),dtype=np.uint8)
  img_z[center_y-h//2:center_y+h//2,center_x-w//2:center_x+w//2]=img_logo
  plt.imshow(img_z)
  plt.show()

img_logo = cv2.imread('logo.jpg')
add_logo(400,250,img_logo)
```

```python
img_logo = cv2.imread('logo.jpg')
add_logo(150,150,img_logo)
```
