```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
!wget -O img.jpg https://github.com/Naxalov/ComputerVision/raw/master/ROAD/road_signs/pedestrian_crossing.jpg
```

```python
img = cv2.imread('img.jpg',cv2.IMREAD_GRAYSCALE)
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#a=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
a=plt.imshow(img,cmap='gray')
print(a)
```

```python
_,b=cv2.threshold(img,175,255,cv2.THRESH_BINARY)
plt.imshow(b,'gray',vmin=0,vmax=255)
plt.show()
```

```python
edge = np.array([
    [-1,-1,-1],
    [-1,8,-1],
    [-1,-1,-1],
])

blur = np.array([
    [1/9,1/9,1/9],
    [1/9,1/9,1/9],
    [1/9,1/9,1/9],
])
```

```python
img1= cv2.filter2D(src=b, ddepth=-1, kernel=edge)
plt.imshow(img1,'gray')
```

```python
cv2_imshow(np.hstack((img1,b)))
```

```python
img1= cv2.imread('img.jpg')
img2=img1[:,:,[2,1,0]]
blur = np.array([
    [1/9,1/9,1/9],
    [1/9,1/9,1/9],
    [1/9,1/9,1/9],
])

blur1=cv2.filter2D(src=img2,ddepth=-1,kernel=blur)
plt.imshow(blur1)
```

```python
cv2_imshow(np.hstack((blur1[:,:,[2,1,0]],img2[:,:,[2,1,0]])))
```
