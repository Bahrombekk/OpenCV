```python
def cat_tom(img,start_point,end_point,color):
  '''`
  Draw tom
  '''
  # CODE
  pst=np.array([[1300,400],
                [1175,400],
                [1175,500]])
  pst1=np.array([[1300,1150],
                 [1175,1150],
                 [1175,1050]])
  pst2=np.array([[1900,400],
                 [2000,400],
                 [2000,500]])
  pst3=np.array([[1950,1150],
                 [2000,1150],
                 [2000,1050]])
  cv.polylines(img,[pst],False,color,thickness=5)
  cv.polylines(img,[pst1],False,color,thickness=5)
  cv.polylines(img,[pst2],False,color,thickness=5)
  cv.polylines(img,[pst3],False,color,thickness=5)
  cv.putText(img,'CAT',(1500,200),16,4,color,thickness=5)
  plt.imshow(img)
  plt.show()
  a=cv.rectangle(img,start_point,end_point,color,thickness=5)

img = cv.imread('cat.jpg')
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
cat_tom(img,(1150,380),(2000,1250),(0, 0, 255))
```

```python
image = np.full((1000,1000,3),255,dtype=np.uint8)
cv.imwrite('image_draw.png',image)

plt.figure(dpi=100)

def draw_home(img):
  '''
  Drawing home
  '''
  pts1=np.array([[800,400],
                 [500,200],
                 [200,400]])
  pts2=np.array([[250,700],
                 [250,400],
                 [750,400],
                 [750,700],
                 [450,700],
                 [450,480],
                 [350,480],
                 [350,700]])
  pts3=np.array([[580,500],
                 [530,500],
                 [530,560],
                 [580,560],
                 [580,500]])
  pts4=np.array([[580,500],
                 [630,500],
                 [630,560],
                 [580,560],
                 [580,500]]) 
  pts5=np.array([[580,560],
                 [530,560],
                 [530,620],
                 [630,620],
                 [630,560]])

  # CODE
  cv.circle(img,(500,500),400,(3, 223, 252),-1)
  cv.polylines(img,[pts1],False,(255,255,255),thickness=40)
  cv.fillPoly(img,[pts2],(255,255,255))
  cv.polylines(img,[pts2],True,(255,255,255),thickness=5)
  cv.fillPoly(img,[pts3],(3, 223, 252))
  cv.polylines(img,[pts3],True,(255,255,255),thickness=15)
  cv.fillPoly(img,[pts4],(3, 223, 252))
  cv.polylines(img,[pts4],False,(255,255,255),thickness=15)
  cv.fillPoly(img,[pts5],(3, 223, 252))
  cv.polylines(img,[pts5],False,(255,255,255),thickness=15)
  cv.line(img,(580,620),(580,560),(255,255,255),thickness=15)
  cv.line(img,(530,560),(630,560),(255,255,255),thickness=15)
  cv.putText(img,'HOME',(350,800),16,3,(255,255,255),thickness=10)
  plt.imshow(img)
  plt.show()

img = cv.imread('image_draw.png')
draw_home(img)
```

```python
image = np.full((1000,1500,3),255,dtype=np.uint8)
cv.imwrite('image_draw.png',image)

plt.figure(dpi=100)
def olympic_circles(img):
  # CODE
  cv.circle(img,(400,400),150,(0,0,255),thickness=20)
  cv.putText(img,'EUROPE',(350,400),16,1,(0,0,255),thickness=5)
  cv.circle(img,(750,400),150,(0,0,0),thickness=20)
  cv.putText(img,'AFRICA',(700,400),16,1,(0,0,0),thickness=5)
  cv.circle(img,(1100,400),150,(255,0,0),thickness=20)
  cv.putText(img,'AMERICA',(1050,400),16,1,(255,0,0),thickness=5)
  cv.circle(img,(570,570),150,(252, 197, 15),thickness=20)
  cv.putText(img,'ASIA',(520,570),16,1,(255,197,15),thickness=5)
  cv.circle(img,(950,570),150,(0,255,0),thickness=20)
  cv.putText(img,'AUSTRALIA',(860,570),16,1,(0,255,0),thickness=5)
  plt.imshow(img)
  plt.show()

img = cv.imread('image_draw.png')
olympic_circles(img)
```

```python
image = np.full((1000,1500,3),0,dtype=np.uint8)
cv.imwrite('image_draw.png',image)

plt.figure(dpi=100)

def pacman(img):
  # CODE
  pts1=np.array([[900,300],
                 [500,500],
                 [900,700]])
  cv.circle(img,(500,500),400,(252, 197, 15),-1)
  cv.circle(img,(600,300),80,(0,0,0),-1)
  cv.circle(img,(630,315),40,(255,255,255),-1)
  cv.fillPoly(img,[pts1],(0,0,0))
  cv.circle(img,(800,500),70,(252, 197, 15),-1)
  cv.circle(img,(1100,500),70,(252, 197, 15),-1)
  cv.circle(img,(1400,500),70,(252, 197, 15),-1)
  cv.putText(img,'PACMAN',(1000,150),16,3,(252, 197, 15),thickness=5)
  plt.imshow(img)
  plt.show()
img = cv.imread('image_draw.png')
pacman(img)
```

```python
image = np.full((1500,1500,3),255,dtype=np.uint8)
cv.imwrite('image_draw.png',image)

plt.figure(dpi=130)

def fir(img):
  # CODE
  pts1=np.array([[750,200],
                 [475,600],
                 [600,600],
                 [250,1000],
                 [370,1000],
                 [50,1400],
                 [1500,1400],
                 [1115,1000],
                 [1250,1000],
                 [875,600],
                 [1000,600],
                 [750,200]])
  cv.polylines(img,[pts1],True,(0,0,255))
  cv.fillPoly(img,[pts1],(0,255,0))
  cv.putText(img,'JUNIPER',(200,200),16,3,(0,255,0),thickness=5)
  plt.imshow(img)
  plt.show()
img = cv.imread('image_draw.png')
fir(img)
```
