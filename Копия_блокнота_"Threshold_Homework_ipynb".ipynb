```python
# utility 
def show(img,b=False):
  plt.figure(figsize=(8,8))#,dpi=d)
  plt.imshow(img,cmap='gray')
  plt.grid(b)
  plt.show()

  
```

```python
def get_images(filename):
  """
  Get image
  Args:
    filename(str): file path
  Returns:
    numpy.ndarray: RGB and Grayscale image
  """
  img=cv2.imread(filename)
  rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  # write code here
  return rgb,gray
```

```python
rgb, gray = get_images('opencv_logo_solid.jpg')
```

```python
gray_org = gray.copy()
```

```python
show(gray)
```

```python
def get_red_threshold(gray):
  """
  threshold image for red
  """
  # write code here
  _,red_image= cv2.threshold(gray,109,255, cv2.THRESH_BINARY)
  return red_image
get_red_threshold(gray)
```

```python
def get_blue_threshold(gray):
  """
  threshold image for blue
  """
  gray[gray<118]=255
  gray[gray>120]=255
  _,blue_image= cv2.threshold(gray,118,255, cv2.THRESH_BINARY)
  # write code here
  return blue_image
get_blue_threshold(gray)
```

```python
def get_green_threshold(gray):
  """
  threshold image for green
  """
  gray_org[gray_org<181]=255
  _,green_image= cv2.threshold(gray_org,181,255, cv2.THRESH_BINARY)
  # write code here
  return green_image
get_green_threshold(gray)
```

```python
get_red = get_red_threshold(gray)
gray = cv2.addWeighted(get_blue_threshold(gray),0.8,get_green_threshold(gray),0.5,0)
```
