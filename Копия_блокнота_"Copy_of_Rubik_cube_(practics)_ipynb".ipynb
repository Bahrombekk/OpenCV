```python
def show(img,s=8):
  plt.figure(figsize=(s,s),dpi=100)
  plt.imshow(img,cmap='gray')
  plt.show()
```

```python
def hsv_opencv(H, S, V):

  H = H//2
  S = S*255//100
  V = V*255//100
  return [H, S, V]
```

```python
def get_img(name):
  img=cv2.imread(name)
  img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
  return img,gray
```

```python
hsv_img=cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
```
