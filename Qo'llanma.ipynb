```python
def windows_logo(img):
  '''
  Drawing windows logo
  '''
  # CODE
  fig=np.array([[0,0],
                [1000,0],
                [1000,1000],
                [0,1000]])
  pts=np.array([[200,200],
                [490,200],
                [490,490],
                [200,490]])
  pts1=np.array([[200,510],
                 [490,510],
                 [490,800],
                 [200,800]])
  pts2=np.array([[510,200],
                 [800,200],
                 [800,490],
                 [510,490]])
  pts3=np.array([[510,510],
                 [800,510],
                 [800,800],
                 [510,800]])
  c=cv2.polylines(img,[fig],True,(0,0,0),thickness=5)
  c1=cv2.fillPoly(c,[fig],(26, 100, 237))
  a=cv2.polylines(img,[pts],True,(0,0,0),thickness=5)
  a1=cv2.polylines(img,[pts1],True,(0,0,0),thickness=5)
  a2=cv2.polylines(img,[pts2],True,(0,0,0),thickness=5)
  a3=cv2.polylines(img,[pts3],True,(0,0,0),thickness=5)
  b1=cv2.fillPoly(a,[pts],(207, 141, 19))
  b2=cv2.fillPoly(a1,[pts1],(22, 161, 29))
  b3=cv2.fillPoly(a2,[pts2],(28, 186, 214))
  b4=cv2.fillPoly(a3,[pts3],(237, 220, 26))
  plt.imshow(img)
  plt.show

img = cv2.imread('image_draw.png')
windows_logo(img)
```

```python
def bounding_box1(img,start_point,end_point,color):
  '''
  Draw bounding box
  '''
  # CODE
  pts=np.array([[200,200],
                [200,800],
                [800,800],
                [800,200]])
  pts1=np.array([[200,200],
                 [240,200],
                 [240,240],
                 [200,240]])
  pts2=np.array([[800,200],
                 [760,200],
                 [760,240],
                 [800,240]])
  pts3=np.array([[200,800],
                 [200,760],
                 [240,760],
                 [240,800]])
  pts4=np.array([[800,800],
                 [760,800],
                 [760,760],
                 [800,760]])
  a=cv2.polylines(img,[pts],True,(0,0,0),thickness=1)
  b=cv2.polylines(img,[pts1],True,(0,0,255),thickness=8)
  b1=cv2.polylines(img,[pts2],True,(0,0,255),thickness=8)
  b2=cv2.polylines(img,[pts3],True,(0,0,255),thickness=8)
  b3=cv2.polylines(img,[pts4],True,(0,0,255),thickness=8)
  c=cv2.fillPoly(img,[pts],(255,255,255))
  plt.imshow(img)
  plt.show()

img = cv2.imread('image_draw.png')
bounding_box1(img,(200,200),(800,800),(0, 0, 255))
```

```python
def bounding_box3(img,start_point,end_point,color):
  '''
  Draw bounding box
  '''
  # CODE
  pts=np.array([[200,200],
                [200,800],
                [800,800],
                [800,200]])
  pts1=np.array([[200,200],
                 [240,200],
                 [240,240],
                 [200,240]])
  pts2=np.array([[800,200],
                 [760,200],
                 [760,240],
                 [800,240]])
  pts3=np.array([[200,800],
                 [200,760],
                 [240,760],
                 [240,800]])
  pts4=np.array([[800,800],
                 [760,800],
                 [760,760],
                 [800,760]])
  a=cv2.polylines(img,[pts],True,(0,0,0),thickness=1)
  b=cv2.polylines(img,[pts1],True,(0,0,0),thickness=3)
  b1=cv2.polylines(img,[pts2],True,(0,0,0),thickness=3)
  b2=cv2.polylines(img,[pts3],True,(0,0,0),thickness=3)
  b3=cv2.polylines(img,[pts4],True,(0,0,0),thickness=3)
  c=cv2.fillPoly(img,[pts],(250, 248, 235))
  c1=cv2.fillPoly(img,[pts1],(255,0,0))
  c2=cv2.fillPoly(img,[pts2],(255,0,0))
  c3=cv2.fillPoly(img,[pts3],(255,0,0))
  c4=cv2.fillPoly(img,[pts4],(255,0,0))
  plt.imshow(img)
  plt.show()

img = cv2.imread('image_draw.png')
bounding_box3(img,(200,200),(800,800),(227, 220, 220))
```

```python

def full_bbox_image(center_x,center_y,width,height,img):
  # Input Format: YOLO
  # CODE
  img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  zeros = np.zeros(img.shape,dtype= np.uint8)
  zeros=cv2.rectangle(zeros,(center_x-width//2,center_y-height//2),(center_x+width//2,center_y+height//2),(255,0,0),-1)
  plt.imshow(cv2.addWeighted(zeros,0.3,img,0.9,0))
  plt.show()

img = cv2.imread('lena.jpg')
full_bbox_image(150,150,100,100,img)
```

```python
def red_threshold(gray):
  _,red_image=cv2.threshold(gray,109,255,cv2.THRESH_BINARY)
  plt.imshow(red_image,cmap='gray')
  plt.show()
  return red_image
red_threshold(gray)
```

```python
def gren_threshold(gray):
  gray[gray<181]=255
  _,gren_image=cv2.threshold(gray,181,255,cv2.THRESH_BINARY)
  plt.imshow(gren_image,cmap='gray')
  plt.show()
  return gren_image
gren_threshold(gray)
```

```python
def get_blue_threshold(gray):
  """
  threshold image for blue
  """
  gray[gray<118]=255
  _,blue_image= cv2.threshold(gray,118,255, cv2.THRESH_BINARY)
  # write code here
  plt.imshow(blue_image,cmap='gray')
  plt.show()
  return blue_image
get_blue_threshold(gray)
     
```
