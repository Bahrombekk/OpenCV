```python
zeros = np.zeros(img.shape,dtype= np.uint8)
```

```python
plt.imshow(img[:,:,[2,1,0]])
plt.show()
```

```python
img[zeros!=0]=zeros[zeros!=0]
```

```python
plt.imshow(img[:,:,[2,1,0]])
plt.show()
```

```python
plt.imshow(img[:,:,[2,1,0]]+zeros[:,:,[2,1,0]])
plt.show()
```
