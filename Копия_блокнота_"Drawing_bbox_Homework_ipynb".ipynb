1. Draw a bbox using points in Pascal VOC format.

<b>Input</b>
  <table align="left"> 
  <tr>
    <th><font size="3">x1</font></th>
    <th><font size="3">y1</font></th>
    <th><font size="3">x2</font></th>
    <th><font size="3">y2</font></th>
  </tr>
  <tr>
    <td><font size="3">1617</font></td>
    <td><font size="3">103</font></td>
    <td><font size="3">1879</font></td>
    <td><font size="3">346</font></td>
  </tr>
  </table>

```python
def bbox_pascal_voc(image, x1, y1, x2, y2):
  """
  Draw a bbox using points in Pascal VOC format.

  Arguments:
    image: numpy array
    x1(int): top left point
    y1(int): top left point
    x2(int): bottom right point
    y2(int): bottom right point
  Returns:
    None: Plot bbox
  """
  a=cv2.rectangle(image,(x1,y1),(x2,y2),(0,255,0),thickness=5)
  plt.imshow(image[:,:,[2,1,0]])
  plt.show()
image=cv2.imread('bus_stop.png')
bbox_pascal_voc(image,1617,103,1879,346)
  # WRITE YOUR CODE HERE
```

```python
image = cv2.imread('bus_stop.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(dpi=150)
# WRITE YOUR CODE HERE
bbox_pascal_voc(image,1617,103,1879,346)
```

```python
def bbox_coco(image, x, y, w, h):
  """
  Draw a bbox using points in COCO format.

  Arguments:
    image: numpy array
    x1(int): top left point
    y1(int): top left point
    w(int): bbox width
    h(int): bbox height
  Returns:
    None: Plot bbox
  """
  a=cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),thickness=5)
  plt.imshow(image[:,:,[2,1,0]])
  plt.show()
image=cv2.imread('bus_stop.png')
bbox_coco(image,244,434,508,238)
  # WRITE YOUR CODE HERE
```
